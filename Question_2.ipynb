```python
import cv2
import numpy as np

def sobel_edge_detection(image):
    # Convert the image to grayscale

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(image, (1, 1), 0)

    # Compute gradients using Sobel operator
    grad_x = cv2.Sobel(blurred, cv2.CV_16S, 1, 0, ksize=1, scale=-1, delta=0, borderType=cv2.BORDER_DEFAULT)
    grad_y = cv2.Sobel(blurred, cv2.CV_16S, 0, 1, ksize=1, scale=-1, delta=0, borderType=cv2.BORDER_DEFAULT)

    # Convert back to uint8
    abs_grad_x = cv2.convertScaleAbs(grad_x)
    abs_grad_y = cv2.convertScaleAbs(grad_y)

    # Combine the gradients to get the final edge-detected image
    edge_detected = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)

    return edge_detected

# Read the input image
input_image = cv2.imread(path)

# Apply Sobel edge detection
edge_image = sobel_edge_detection(input_image)

# Display the input and edge-detected images
cv2_imshow(input_image)
cv2_imshow(edge_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
